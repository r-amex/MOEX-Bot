```python
import pandas as pd
```

```python
df = pd.read_csv('D:/Work/Data/RIU2_5min.csv', delimiter=';', header = 0)
```

```python
oi = pd.read_csv('D:/Work/Data/OI_RIU2.csv', delimiter=';', header = 0)
```

```python
# переименовываем колонку
df.rename(columns={'Datetime': 'Trade_Time'}, inplace=True)
```

```python
# преобразуем Trade_Time из строки в формат datetime 
oi['Trade_Time']= pd.to_datetime(oi['Trade_Time'])
df['Trade_Time']= pd.to_datetime(df['Trade_Time'])
```

```python
# объединяем два файла
#od = pd.merge(df, oi, how = 'left', on=['Trade_Time'])
od = pd.merge(df, oi, how = 'left')
```

```python
# Сортировка по возрастанию ( день ближе к сегодня - последний)
#od = od.sort_values(by='Trade_Time', ascending=True, na_position='first')
```

```python
# заполняем NaN предыдущими значениями
out = od.ffill(axis= 0)
```

```python
# Сохраняем все в файл
out.to_csv('D:/Work/Data/RIU2_5min_OI.csv',
          index= False,
          sep = ";")
```
